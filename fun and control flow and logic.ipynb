```python
Number Guessing Game:
The Goal: The program randomly selects a number between 1 and 100.
    The user guesses the number, and the program replies with "Too high!", "Too low!", or "You got it!".
        The game loops until the user guesses correctly.
Concepts Tested: while loops, if/elif/else statements, comparison operators, importing the basic random module.
```

```python
import random  
random_number = random.randint(1, 100) 
while True: 
    guess = int(input("Guess a number between 1 and 100: "))
    if guess > random_number:
        print("Too high!")
    elif guess < random_number:
        print("Too low!")
    else:
        print("You guess it right!")
        break 
```

```python
Rock, Paper, Scissors:
The Goal: A continuous loop where the user plays Rock, Paper, Scissors against the computer. Keep track of the score until the user types "quit".
Concepts Tested: Infinite loops with break conditions, complex if/elif logic, string normalization (e.g., using .lower()).
```

```python
import random

user_score = 0
computer_score = 0

while True:
    user_answer = input("Enter Rock, Paper, Scissors (or quit): ").lower()
    if user_answer == "quit":
        break
    if user_answer not in ["rock", "paper", "scissors"]:
        print("The input is not correct")
        continue

    computer = random.choice(["rock", "paper", "scissors"])
    print("Computer:", computer)

    if (user_answer == "rock" and computer == "scissors") or \
       (user_answer == "paper" and computer == "rock") or \
       (user_answer == "scissors" and computer == "paper"):
        print("You win!") 
        user_score += 1
    elif (user_answer == "scissors" and computer == "rock") or \
         (user_answer == "rock" and computer == "paper") or \
         (user_answer == "paper" and computer == "scissors"):
        print("You lost!")
        computer_score += 1
    else:
        print("Draw!")

    print("Score:", user_score, "-", computer_score)
```

```python
The Virtual Barista (Simple Point-of-Sale)
The Goal: Display a text menu of three coffee shop drinks with different prices. 
    Create a continuous loop that asks the user what they want to order. 
        Add the price of their chosen drink to a running total. 
            If they type an invalid drink, tell them it's not on the menu. 
            The loop should continue until the user types "checkout", at which point the program prints their final total.
Concepts Tested: Infinite while loops, break conditions, accumulating totals (+= operator), 
complex if/elif/else logic, and string normalization (using .lower() to handle if the user types "Checkout" or "CHECKOUT").
```

```python
print("Hello bong! what do you want to drink?") 
print("menu of three coffee shop drinks with different prices:")
print("Coffee=$1.00")
print("Matcha=$1.50")
print("Milk=$0.50")
total = 0
while True:
    answer = input("Enter your order (or 'checkout'): ").lower().strip()
    if answer == "checkout":
        break
    elif answer == "coffee":
        total += 1.00
        print("Added Coffee")
    elif answer == "matcha":
        total += 1.50
        print("Added Matcha")
    elif answer == "milk":
        total += 0.50
        print("Added Milk")
    else:
        print("Sorry bong, that’s not on the menu.")
print(f"Your total is: ${round(total, 2):.2f}")
```

```python
The Mental Math Challenger
The Goal: The program generates two random numbers between 1 and 12 and asks the user to multiply them together (e.g., "What is 7 x 8?"). 
    If the user gets it right, they get a point, and the program asks another question. 
        If they get it wrong, the program tells them the correct answer and ends the game, displaying their final streak/score.
Concepts Tested: Importing the random module, tracking score variables, while loops dependent on a boolean condition (e.g., while is_correct:),
if/else statements, and type casting (converting the user's answer to an int to compare it to the actual math result).
```

```python
import random
score = 0
is_correct = True
while is_correct:
    random_num1 = random.randint(1, 12)
    random_num2 = random.randint(1, 12)
    answer = int(input(f"What is {random_num1} x {random_num2}? "))
    correct_answer = random_num1 * random_num2
    if answer == correct_answer:
        print("The answer is correct")
        score += 1
    else:
        print("The answer is wrong")
        print(f"The correct answer is: {correct_answer}")
        is_correct = False
print(f"Your score is: {score}")
```
